```python
# -------------------------------------------------
# Imports
# -------------------------------------------------
import pandas as pd                 # data handling
import matplotlib.pyplot as plt     # core plotting
import seaborn as sns               # high‑level visualisation API

sns.set(style="whitegrid")          # optional – nicer default style


# -------------------------------------------------
# Load the Iris dataset
# -------------------------------------------------
df = pd.read_csv("/content/IRIS.csv")
```

```python
# ----------------------------------------------
# Quick inspection of the data
# ------------------------------------------
print("Dataset shape (rows, cols):", df.shape)          # e.g. (150, 6)
print("Column names:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
```

```python
# -------------------------------------------------
# Visualisations
# -------------------------------------------------

# a) Scatter plot: choose any two numeric columns
plt.figure(figsize=(7, 5))
sns.scatterplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    hue="species",
    palette="deep"
)
plt.title("Sepal Length vs. Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.tight_layout()
plt.show()
```

```python
# b) Histogram: distribution of a single numeric column
plt.figure(figsize=(7, 4))
sns.histplot(
    data=df,
    x="petal_width",
    bins=15,
    kde=True,
    color="steelblue"
)
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# c) Box plot: spread & outliers per species (or any categorical column)
plt.figure(figsize=(7, 5))
sns.boxplot(
    data=df,
    x="species",
    y="petal_width",
)
plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.tight_layout()
plt.show()
```
